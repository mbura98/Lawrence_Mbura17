```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data.csv')
data.drop('Unnamed: 0', axis=1, inplace=True)
data.head()
```

```python
data.head()
```

```python
#Minimum price of an item in the dataset
data['Price'].min()
```

```python
#Maximum price of an item in the dataset
data['Price'].max()
```

```python
#Total number of transactions done
data['Transaction_ID'].count()
```

```python
#mean price of all items sold to 2 dp
data['Price'].mean().round(2)
```

```python
#Total sum of items sold
data['Total_Sale_Amount'].sum()
```

```python
#Total number of items sold
data['Quantity'].sum()
```

```python
#Mean rating of items to 2dp
data['Customer_rating'].mean().round(2)
```

```python
#Number of counts for product sold
data['Product_Category'].value_counts()
```

```python
#Number of times discount was awarded
data['Discount_Code'].value_counts()
```

```python
#Total amount generated from the data
data['Total_Sale_Amount'].sum()
```

```python
#Total sum of items based on their product category
data.groupby('Product_Category')['Price'].sum()
```

```python
#Average mean price of each product
data.groupby('Product_Category')['Price'].mean()
```

```python
#Mean rating of product category
data.groupby('Product_Category')['Customer_rating'].mean().round(2)
```

```python
#Total sum of quantity of product category sold
data.groupby('Product_Category')['Quantity'].sum()
```
