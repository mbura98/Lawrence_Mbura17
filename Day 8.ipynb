```python
weather_data=pd.read_csv('seattle-weather.csv')
weather_data.head()
```

```python
weather_data['date']=pd.to_datetime(weather_data['date'])
```

```python
weather_data.duplicated().sum()
```

```python
weather_data.isnull().sum()
```

```python
# Set a style for the plots
sns.set_theme(style="whitegrid")

# Identify the columns to plot
numerical_cols = ['precipitation', 'temp_max', 'temp_min', 'wind']

# Create subplots
fig, axes = plt.subplots(nrows=len(numerical_cols), ncols=1, figsize=(14, 18), sharex=True)
plt.suptitle('Line Plots of Seattle Weather Variables (2012-2015)', fontsize=16, y=1.0)

for i, col in enumerate(numerical_cols):
    sns.lineplot(weather_data, x=weather_data.index, y=col, ax=axes[i], color='teal', linewidth=1)
    axes[i].set_title(f'{col.replace("_", " ").title()} Over Time', fontsize=14)
    axes[i].set_ylabel(col.replace("_", " ").title())
    axes[i].grid(True)

axes[-1].set_xlabel("Date")
plt.tight_layout(rect=[0, 0, 1, 0.98])
plt.show()
```

```python
from scipy import stats

# Calculate the 30-day rolling mean for smoothing the daily variations
rolling_window = 30
weather_data_rolling = weather_data[numerical_cols].rolling(window=rolling_window, center=True).mean()

print('\nLong-Term Trend Analysis (Linear Regression)')

for col in numerical_cols:
    # Prepare data for linear regression: using a numerical representation of time
    # This assumes a continuous, linear time trend from the start of the dataset
    weather_data_temp = weather_data.dropna(subset=[col])
    x = np.arange(len(weather_data_temp))
    y = weather_data_temp[col].values

    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

    # Interpretation
    if p_value < 0.05:
        significance = 'Statistically Significant'
        trend_direction = 'Increasing' if slope > 0 else 'Decreasing'
    else:
        significance = 'Not Statistically Significant'
        trend_direction = 'No Strong Linear'

    print(f'\n**{col.replace('_', ' ').title()}**')
    print(f'Slope (Trend Rate): {slope:.4f}')
    print(f'P-value: {p_value:.4f}')
    print(f'R-squared: {r_value**2:.4f}')
    print(f'Conclusion: The long-term trend is **{trend_direction}** ({significance}).')

    # Plot the rolling mean and the fitted line
    plt.figure(figsize=(14, 4))
    sns.lineplot(weather_data_rolling, x=weather_data_rolling.index, y=col, label=f'{rolling_window}-Day Rolling Mean', color='teal', linewidth=2)
    plt.plot(weather_data_temp.index, intercept + slope * x, color='red', linestyle='--', label=f'Linear Trend Line (Slope: {slope:.4f})')
    plt.title(f'Trend Analysis for {col.replace("_", " ").title()}')
    plt.xlabel('Date')
    plt.ylabel(col.replace('_', ' ').title())
    plt.legend()
    plt.show()
```
