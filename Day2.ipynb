```python
#Importing the needed libraries
import pandas as pd
import numpy as np
```

```python
#DataFrame for subjects and marks
data={
    'Subject': [
        'Mathematics',
        'English',
        'Chemistry',
        'Computer Science',
        'History',
        'Physics'
    ],
    'Marks':[
        85,
        92,
        78,
        95,
        88,
        72
    ]
}
#create the DataFrame
marks_df=pd.DataFrame(data)
print(marks_df) #Display the DataFrame
```

```python
#.head() displays the first 5 rows of the dataframe by default
marks_df.head()
```

```python
#.tail() it displays the last 5 rows of the dataframe by default
marks_df.tail(3)#displa the last 3 rows
```

```python
#Provides a summary of data types, non-null values and memory usage
marks_df.info()
```

```python
#Returns a tuple of rows and columns
marks_df.shape
```

```python
#Returns the data type for each column
marks_df.dtypes
```

Basic Selection and Slicing

```python
#Selecting a single column
marks_df['Marks']#selects the marks column
```

```python
#Selecting multiple columns
marks_df[['Subject', 'Marks']]
```

```python
#Selecting rows using loc(Label-based indexing) eg to display the second item
row_by_label = marks_df.loc[1]
row_by_label
```

```python
#Selecting rows using iloc(Integer position indexing)
row_by_position = marks_df.iloc[0]
row_by_position
```
