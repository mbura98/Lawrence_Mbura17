```python
def distplots(col):
    sns.distplot(df[col])
    plt.show()
```

```python
for i in list(df.select_dtypes(exclude=['object']).columns):
    distplots(i)
```

```python
df.groupby('Gender')[['Annual Income (k$)','Spending Score (1-100)']].value_counts()
```

```python
plt.figure(figsize=(20,8))
sns.countplot(df,x='Annual Income (k$)',hue='Gender')
plt.title('Annual Income Based On Gender')
```

```python
plt.figure(figsize=(20,8))
sns.countplot(df,x='Spending Score (1-100)',hue='Gender')
plt.title('Spending Score Based On Gender')
```

```python
count=df['Gender'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(count,labels=count.index,autopct='%1.1f%%',startangle=140)
plt.title('Spending Score Based On Gender')
```
