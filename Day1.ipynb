```python
LISTS
-They are denoted by []
-They are mutable(content can be changed)
-They can contain various data type
```

```python
my_list=[1, 'apple', 3.14, True]
```

```python
#Accessing output apple
my_list[1]
```

```python
#slicing example
my_list[0:2]
#It access items from 0 t0 2 not including the 2 index value
```

```python
#Adding an item in the list
my_list.append('VLC')
print(my_list)
#It adds string VLC into the list 
```

```python
#finding the length of my list
len(my_list)
```

```python
DICT
-They are denoted by {}
-They are mutable
-They store data as pairs
-They maintain the order they are stored
```

```python
my_dict={'name': 'Peter', 'age': 30, 'city': 'Nairobi'}
```

```python
#Accessing Value eg age in the dict.
my_dict['age']
```

```python
#Adding on the dict
my_dict['year']= 1964
my_dict['model']= 'BMW'
#checking the updated dict
my_dict
```

```python
#updating my_dict
my_dict['age']= 25
```

```python
#Removing an item on the dict
my_dict.pop('model')
```

```python
#Getting keys on the dict
my_dict.keys()
```

```python
#Getting values on the dict
my_dict.values()
```
