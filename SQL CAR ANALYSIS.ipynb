TOP 5 EXPENSIVE CARS BASED ON NAME, BRAND, MODEL, YEAR OF MANUFACTURE AND FUEL TYPE

```python
%sql select max(price) as car_price, name, brand, model, year, fuel\
from CRS\
group by name\
order by max(price) desc limit 5;
```

5 LEAST EXPENSIVE CARS BASED ON NAME, BRAND, MODEL, YEAR OF MANUFACTURE AND FUEL TYPE

```python
%sql select min(price) as car_price, name, brand, model, year, fuel\
from CRS\
group by name\
order by min(price) asc limit 5;
```

TOP 7 UNIQUE NAME DIESEL CARS

```python
%sql select distinct name as Diesel_Cars from CRS where fuel=='Diesel' limit 7;
```

TOP 7 UNIQUE NAME OF CARS THAT HAVE CC GREATER THAN 2400

```python
%sql select distinct name as T4CC_Cars from CRS where CC>2400 limit 7;
```
