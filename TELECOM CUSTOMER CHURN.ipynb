```python
df.drop('customerID', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
enco=LabelEncoder()
df['Churn']=enco.fit_transform(df['Churn'])
df['gender']=enco.fit_transform(df['gender'])
df['SeniorCitizen']=enco.fit_transform(df['SeniorCitizen'])
df['Partner']=enco.fit_transform(df['Partner'])
df['Dependents']=enco.fit_transform(df['Dependents'])
df['PhoneService']=enco.fit_transform(df['PhoneService'])
df['MultipleLines']=enco.fit_transform(df['MultipleLines'])
df['InternetService']=enco.fit_transform(df['InternetService'])
df['OnlineSecurity']=enco.fit_transform(df['OnlineSecurity'])
df['OnlineBackup']=enco.fit_transform(df['OnlineBackup'])
df['DeviceProtection']=enco.fit_transform(df['DeviceProtection'])
df['TechSupport']=enco.fit_transform(df['TechSupport'])
df['StreamingTV']=enco.fit_transform(df['StreamingTV'])
df['StreamingMovies']=enco.fit_transform(df['StreamingMovies'])
df['Contract']=enco.fit_transform(df['Contract'])
df['PaperlessBilling']=enco.fit_transform(df['PaperlessBilling'])
df['PaymentMethod']=enco.fit_transform(df['PaymentMethod'])
df['MonthlyCharges']=enco.fit_transform(df['MonthlyCharges'])
df['TotalCharges']=enco.fit_transform(df['TotalCharges'])
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
x=df.drop('Churn', axis=1)
y=df['Churn']
```

```python
from sklearn.preprocessing import StandardScaler
sca=StandardScaler()
x=sca.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lreg=LogisticRegression()
lreg.fit(x_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
```

```python
yhat_train_lreg=lreg.predict(x_train)
yhat_test_lreg=lreg.predict(x_test)
```

```python
print(classification_report(y_train, yhat_train_lreg))
print()
print(classification_report(y_test,yhat_test_lreg))
```

```python
print(confusion_matrix(y_train, yhat_train_lreg))
print()
print(confusion_matrix(y_test, yhat_test_lreg))
```

```python
c_matrix=ConfusionMatrixDisplay(confusion_matrix(yhat_train_lreg, y_train))
c_matrix.plot(cmap='coolwarm')
```

```python
c_matrix=ConfusionMatrixDisplay(confusion_matrix(yhat_test_lreg, y_test))
c_matrix.plot(cmap='viridis')
```

```python
from sklearn.model_selection import cross_val_score
scores=cross_val_score(lreg, x, y, cv=3)
np.mean(scores)
```

RANDOM FOREST CLASSIFIER MODEL

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train, y_train)
```

```python
yhat_train_rfc=rfc.predict(x_train)
yhat_test_rfc=rfc.predict(x_test)
```

```python
print(classification_report(y_train, yhat_train_rfc))
print()
print(classification_report(y_test, yhat_test_rfc))
```

```python
print(confusion_matrix(y_train,yhat_train_rfc))
print()
print(confusion_matrix(y_test,yhat_test_rfc))
```

```python
c_matrix=ConfusionMatrixDisplay(confusion_matrix(yhat_train_rfc, y_train))
c_matrix.plot(cmap='coolwarm')
plt.show()
```

```python
c_matrix=ConfusionMatrixDisplay(confusion_matrix(yhat_test_rfc, y_test))
c_matrix.plot(cmap='viridis')
plt.show()
```

```python
scores=cross_val_score(rfc, x, y, cv=3)
np.mean(scores)
```
