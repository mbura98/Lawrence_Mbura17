```python
from sklearn.preprocessing import LabelEncoder
enco=LabelEncoder()
data['Brand']=enco.fit_transform(data['Brand'])
data['Model']=enco.fit_transform(data['Model'])
data['Fuel_Type']=enco.fit_transform(data['Fuel_Type'])
data['Transmission']=enco.fit_transform(data['Transmission'])
data['Owner_Type']=enco.fit_transform(data['Owner_Type'])
```

```python
x=data.drop(['Price'],axis=1)
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scala=StandardScaler()
x_train=scala.fit_transform(x_train)
x_test=scala.fit_transform(x_test)
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data.corr(),annot=True,cmap='coolwarm')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg.fit(x_train,y_train)
```

```python
yhat=lreg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
print("Accuracy: ",r2_score(y_test,yhat))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
```

```python
yhat2=rfc.predict(x_test)
```

```python
print("Accuracy: ",r2_score(y_test,yhat2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
yhat6=knn.predict(x_test)
print("Accuracy :",r2_score(y_test,yhat6))
```
