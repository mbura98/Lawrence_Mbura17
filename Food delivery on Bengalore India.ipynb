```python
#IMPORTING FOLIUM FOR CREATION OF LEAFLET MAPS
import folium
```

```python
#CREATING A MAP OF BANGOLORE CITY
food_map = folium.Map(location=[12.9716, 77.5946], zoom_start=12)
food_map
```

```python
x=df['latitude']
y=df['longitude']
```

```python
#DATAFRAME OF LATITUDE AND LATITUDE COLUMN AND NAMING IT LOCATIONS
locations=pd.DataFrame({'latitude':x, 'longitude':y})
```

```python
#ADDING THE LOCATIONS OF WHERE FOOD WAS ORDERED FROM INTO BANGALORE MAP
for index, location in locations.iterrows():
    folium.Marker([location['latitude'], location['longitude']]).add_to(food_map)
```
