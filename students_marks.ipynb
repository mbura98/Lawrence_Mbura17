```python
sum_columns=['sql_marks', 'excel_marks', 'python_marks', 'power_bi_marks', 'english_marks']
df['total_marks']=df[sum_columns].sum(axis=1)
```

```python
df.head()
```

```python
df['rank']=df['total_marks'].rank(method='min', ascending=False).astype(int)
df.head()
```

```python
df['location'].value_counts()
```

```python
top_5=df[df['rank']<=5]
top_5
```

```python
top_10=df.sort_values(by='rank', ascending=True).head(10)
top_10
```
