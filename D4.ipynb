```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('data.csv')
data.head()
```

```python
#drop the unnecessary 'Unnamed: 0' column
data.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
#create Total Sale Amount column
data['Total_Sale_Amount']=data['Price'] * data['Quantity']
```

```python
data['Weighted_Rating'] = data['Customer_rating'] * data['Quantity']
#This feature combines the rating with the volume(quantity) of the transaction
#A high rating on a large quantity purchase might be more significant than a high rating on a single cheap item
```

```python
#Creating Price Tier Bins
bins=[0, 10, 50, data['Price'].max() + 1]
labels= ['Low_Price', 'Mid_Price', 'High_Price']
data['Price_Tier'] = pd.cut(data['Price'], bins=bins, labels=labels, right=False)

#Group the continous Price column into 'Low', 'Medium', 'High' buckets 
#Low_Price:[0,10] - Price >= 0 and <1 0
#Mid_Price:[10,49] - Price >= 10 and < 49
#High_Price:[49,max] - Price >= 49 and Maximum Price in Data
#right=False means the interval is left inclusive(price is included) ansd right exclusive(upper boundary is not included)
```
